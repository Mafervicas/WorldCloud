```python
#Importamos la librer√≠a
from newspaper import Article
```

```python
#Guardamos todo lo de la pagina en una variable
pagina = Article('https://fococonsultores.es/las-cualidades-de-un-buen-jefe/')
pagina.download()
pagina.parse()
```

```python
#Nos quedamos con el texto
pagina.text
```

```python
#Importamos librer√≠as
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from wordcloud import STOPWORDS
```

```python
imagen = WordCloud()
imagen.generate(pagina.text)
plt.imshow(imagen, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
#Importamos librer√≠a
from PIL import Image
import numpy as np
```

```python
mask = np.array(Image.open('C:\\Users\\mafer\\Documents\\Programas.Maf\\PythonScripts\\DataAnalitics\\Maf/trofeo.png'))
wc = WordCloud(stopwords=STOPWORDS,
               mask=mask, background_color="white",min_word_length=4 ,contour_width=3, contour_color='green',
               max_words=2500, max_font_size=256,
               random_state=42)
wc.generate(pagina.text)
plt.figure( figsize=(25,20) )
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```
